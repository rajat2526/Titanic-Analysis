# Put Dummies Values

```python
Sex=pd.get_dummies(titanic_data['Sex'],drop_first=True)      #Put Dummies Vlues in 'Sex' Column and drop Extra first column.
Sex.head(5)                                                  
```

```python
Embarked=pd.get_dummies(titanic_data['Embarked'],drop_first=True)       #Put Dummies Vlues in 'Embarked' Column and drop Extra first column.
Embarked.head(5)
```

```python
Pclass=pd.get_dummies(titanic_data['Pclass'],drop_first=True)     #Put Dummies Vlues in 'Pclass' Column and drop Extra first column.
Pclass.head(5)
```

```python
titanic_data=pd.concat([titanic_data,Sex,Embarked,Pclass],axis=1)   #Now Concat dummeis values columns.
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Sex','Embarked','PassengerId','Name','Ticket','Pclass'],axis=1,inplace=True)     #Drop Original columns.
```

```python
titanic_data.head(5)
```

# Train Data

```python
x= titanic_data.drop(["Survived"],axis=1)
y= titanic_data["Survived"]
```

```python
 from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression     #import logisticRegression.
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)         #Apply LogisticRegression.
```

# Test Model

```python
prediction= logmodel.predict(X_test)     #Now test our train Model.
```

```python
from sklearn.metrics import classification_report       #create Classification Report.
```

```python
classification_report(y_test,prediction)         
```

```python
from sklearn.metrics import confusion_matrix         #Show Accuracy through confusion_Matri..
```

```python
confusion_matrix(y_test,prediction)                  #Show Confusion_Matrix.
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)             #Show Accuracy_Score.
```
